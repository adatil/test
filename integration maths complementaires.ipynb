# Méthodes d'approximations d'intégrales (aire sous la parabole)

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
def f(x):
    return x**2
```

## Méthode des rectangles

```python
# integration numerique par la methode des rectangles (somme inférieure)

xmin = 0
xmax = 1
nbx = 20
nbi = nbx - 1 # nombre d'intervalles

x = np.linspace(xmin, xmax, nbx)
y = f(x)
plt.plot(x,y,"bo-")

integraleinf = 0
for i in range(nbi):
    integraleinf = integraleinf + y[i]*(x[i+1]-x[i])
    # dessin du rectangle
    x_rect = [x[i], x[i], x[i+1], x[i+1], x[i]] # abscisses des sommets
    y_rect = [0   , y[i], y[i]  , 0     , 0   ] # ordonnees des sommets
    plt.plot(x_rect, y_rect,"r")
print("integrale >=", integraleinf)
plt.show()
```

```python
# integration numerique par la methode des rectangles (somme supérieure)

xmin = 0
xmax = 1
nbx = 20
nbi = nbx - 1 # nombre d'intervalles

x = np.linspace(xmin, xmax, nbx)
y = f(x)
plt.plot(x,y,"bo-")

integralesup = 0
for i in range(nbi):
    integralesup = integralesup + y[i+1]*(x[i+1]-x[i])
    # dessin du rectangle
    x_rect = [x[i], x[i], x[i+1], x[i+1], x[i]] # abscisses des sommets
    y_rect = [0   , y[i+1], y[i+1]  , 0     , 0   ] # ordonnees des sommets
    plt.plot(x_rect, y_rect,"r")
print("integrale <=", integralesup)
plt.show()
```

```python
print("on peut donc dire que l'aire cherchée est comprise entre :", integraleinf," et ",integralesup)
```

## Méthode des trapèzes

```python
# integration numerique par la methode des trapèzes 

xmin = 0
xmax = 1
nbx = 20
nbi = nbx - 1 # nombre d'intervalles

x = np.linspace(xmin, xmax, nbx)
y = f(x)
plt.plot(x,y,"bo-")

integralesup = 0
for i in range(nbi):
    integralesup = integralesup + (y[i+1]+y[i])/2*(x[i+1]-x[i])
    # dessin du trapèze
    x_rect = [x[i], x[i], x[i+1], x[i+1], x[i]] # abscisses des sommets
    y_rect = [0   , y[i], y[i+1]  , 0     , 0   ] # ordonnees des sommets
    plt.plot(x_rect, y_rect,"r")
print("integrale <=", integralesup)
plt.show()
```

## Méthode de Monte Carlo

```python
N = 10000

# On tire un grand nombre de points uniformément sur [0,1]^2
x = np.random.random(size=N)
y = np.random.random(size=N)

# On créer un array de booléens indiquant si le point se situe en dessous de la courbe
z = (y)<f(x)
```

```python
# plt.subplots() permet de créer un objet figure des objets axes.
fig, ax = plt.subplots(1,1,figsize=(5,5))

# L'indexage booléen fait partie des "indexages avancés" de numpy.
# Il est possible de filter un array en l'indexant par un autre array contenant 
# des booléens. On obtient un array plus petit ne contenant que les elements aux emplacements "True".
plt.scatter(x[z], y[z], marker='.')

# Logical not permet d'inverser la valeur logique des éléments d'un np.array.
nz = np.logical_not(z) 
plt.scatter(x[nz], y[nz], marker='.')

# La méthode set de l'objet `axe` permet de spécifier le titre. 
ax.set(title=r"Aire sous la courbe par la méthode de Monte-Carlo", 
       xlabel='x',
       ylabel='y')

# La méthode savefig de l'objet `figure` permet de sauvegarder l'image dans un fichier.
# Le format est déduit de l'extension. 
fig.savefig('monte_carlo.png')
```

```python
# On estime l'aire
aire =z.sum()/N
print("Estimation de l'aire : {}".format(aire))
```
